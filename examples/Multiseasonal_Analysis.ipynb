```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./PrimarySamplingGRES.csv')
data.head()
```

```python
# Преобразуем столбец с датами в datetime, если еще не сделано
data['Date'] = pd.to_datetime(data['Date'])

# Создаем полный диапазон дат
min_date = data['Date'].min()
max_date = data['Date'].max()
full_dates = pd.date_range(min_date, max_date, freq='h')  # 'H' для часового интервала

# Создаем полный DataFrame и соединяем с данными
full_data = data.set_index('Date').reindex(full_dates)

# Восстанавливаем имя индекса
full_data.index.name = 'Date'
full_data
```

```python
colTurb = ['TurbineUnitPower1',
           'TurbineUnitPower2',
           'TurbineUnitPower3',
           'TurbineUnitPower4',
           'TurbineUnitPower5',
           'TurbineUnitPower6',
           'TurbineUnitPower7',
           'TurbineUnitPower8']

colWater = ['TurbineTempFeedWater1',
            'TurbineTempFeedWater2',
            'TurbineTempFeedWater3',
            'TurbineTempFeedWater4',
            'TurbineTempFeedWater7',
            'TurbineTempFeedWater5',
            'TurbineTempFeedWater6',
            'TurbineTempFeedWater8']

coldrop = ['BoilerTempColdAirLeft1', 'BoilerTempColdAirLeft2',
           'BoilerTempColdAirLeft3', 'BoilerTempColdAirLeft4',
           'BoilerTempColdAirLeft5', 'BoilerTempColdAirLeft6',
           'BoilerTempColdAirLeft7', 'BoilerTempColdAirLeft8',
           'BoilerTempColdAirLeft9', 'BoilerTempColdAirLeft10',
           'BoilerTempColdAirLeft11', 'BoilerTempColdAirRight1',
           'BoilerTempColdAirRight2', 'BoilerTempColdAirRight3',
           'BoilerTempColdAirRight4', 'BoilerTempColdAirRight5',
           'BoilerTempColdAirRight6', 'BoilerTempColdAirRight7',
           'BoilerTempColdAirRight8', 'BoilerTempColdAirRight9',
           'BoilerTempColdAirRight10', 'BoilerTempColdAirRight11',
           'StationConsumpMasut', 'StationConsumpOTPKaltan1',
           'StationConsumpOTPHP', 'StationConsumpOTPOsinniki',
           'StationConsumpOTPTHF', 'StationConsumpOTPKVOiT',
           'StationConsumpElectricityTurbineUnit'
]

df = (
    full_data
    .assign(TurbineTempFeedWaterQ2 = lambda x: x[colWater].quantile(q = 0.5, axis = 1))
    .assign(TurbinePowerSum = lambda x: x[colTurb].sum(axis=1))
    .drop(columns = colTurb + colWater + coldrop)
)

df.loc[df['TurbinePowerSum'] == 0, 'TurbinePowerSum'] = np.nan
df
```

```python
df.info()
```

```python
df = df.resample("D").mean()
df = df.sort_index()
df
```

```python
df['Year'] = df.index.year
df = df[(df['Year'] > 2007) & (df['Year'] < 2016)]
```

```python
def month_to_season(month):
    if month in [12, 1, 2, 3]:
        return 1  # Зима
    elif month in [4, 5]:
        return 2  # Весна
    elif month in [6, 7, 8]:
        return 3  # Лето
    elif month in [9, 10, 11]:
        return 4  # Осень

def add_time_features(df):
    df = df.copy()
    df['Year'] = df.index.year
    df['Month'] = df.index.month
    df['Day'] = df.index.day
    df['DayOfYear'] = df.index.dayofyear
    df['DayOfWeek'] = df.index.dayofweek + 1
    df['Season'] = df.index.month.map(month_to_season)
    return df

df = add_time_features(df)
df
```

```python
print(df.dtypes)
```

```python
col = ['Year', 'Season', 'Month', 'DayOfYear', 'Day', 'DayOfWeek']
df[col] = df[col].astype('int')
df.info()
```

## SARIMAX

```python
df.info()
```

```python
target = 'TurbinePowerSum'

features = [
    'StationTempOutdoorAir',
    'TurbineTempFeedWaterQ2',
    'StationCoalHumidity',
    'StationCoalAsh',
    'StationConsumpNaturalFuel',
    'Year',
    'Season',
    'DayOfWeek'
]
```

```python
df_day = df[[target]+ features].resample("D").mean()
df_day
```

```python
train_day = df_day.loc[df_day.index < '2015-01-01']
test_day = df_day.loc[df_day.index >= '2015-01-01']
```

```python
import statsmodels.api as sm

mod = sm.tsa.statespace.SARIMAX(
    endog = train_day[target],
    exog = train_day[features],
    trend='c',
    order=(0,0,0),
    freq = 'D'
)

res = mod.fit(disp=False)
print(res.summary())
```

```python
pred_day = res.forecast(test_day.shape[0], exog = test_day[features])
```

```python
pred_day
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
print('R2 score:', r2_score(test_day[target], pred_day))
print('MAE score:', mean_absolute_error(test_day[target], pred_day))
print('MSE score:', mean_squared_error(test_day[target], pred_day))
print('RMSE score:', np.sqrt(mean_squared_error(test_day[target], pred_day)))
print('MAPE score:', mean_absolute_percentage_error(test_day[target], pred_day))
```

```python
result = sm.tsa.seasonal_decompose(df_day[target].resample("M").mean(), model = 'additive')
trend = result.trend.dropna()
seasonal = result.seasonal.dropna()
residual = result.resid.dropna()

# Plot the decomposed components
plt.figure(figsize=(6,6))

plt.subplot(4, 1, 1)
plt.plot(df_day[target], label='Original Series')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(trend, label='Trend')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(seasonal, label='Seasonal')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(residual, label='Residuals')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
col_df = ['TurbinePowerSum',
          'StationTempOutdoorAir',
          'TurbineTempFeedWaterQ2',
          'StationCoalHumidity',
          'StationCoalAsh',
          'StationConsumpNaturalFuel',
          'Year',
          'Season',
          'Month',
          'DayOfYear',
          'Day',
          'DayOfWeek'
          ]

df_cat = df[col_df]
df_cat
```

```python
target = 'TurbinePowerSum'

features = [
    'StationTempOutdoorAir',
    'TurbineTempFeedWaterQ2',
    'StationCoalHumidity',
    'StationCoalAsh',
    'StationConsumpNaturalFuel',
    'Year',
    'Season',
    'Month',
    'DayOfYear',
    'Day',
    'DayOfWeek'
]
```

```python
train_cat = df_cat.loc[df_cat.index<'2015-01-01']
test_cat = df_cat.loc[df_cat.index>='2015-01-01']

X_train = train_cat[features]
y_train = train_cat[target]

X_test = test_cat[features]
y_test = test_cat[target]
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.1,
    depth=6,
    loss_function='RMSE',
    random_seed=42,
    early_stopping_rounds=50,
    task_type='CPU'
)

model.fit(
    X_train, y_train,
    eval_set=(X_test, y_test),
    cat_features=col_cat,
    verbose=100
)
```

```python
pred_cat = pd.Series(data = model.predict(data = X_test), index = X_test.index)
```

```python
print(model.get_params())
print(model.get_feature_importance())
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
print('R2 score:', r2_score(test_cat[target], pred_cat))
print('MAE score:', mean_absolute_error(test_cat[target], pred_cat))
print('MSE score:', mean_squared_error(test_cat[target], pred_cat))
print('RMSE score:', np.sqrt(mean_squared_error(test_cat[target], pred_cat)))
print('MAPE score:', mean_absolute_percentage_error(test_cat[target], pred_cat))
```
